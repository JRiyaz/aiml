```python
# 1. Add the 1s to X
# 2. Initialze random weights
# 3. Start with perceptron - visit every point
# 4. Calculate y predicted for every point
# 5. check for misclassification
# 6. Update weight if applicable
# 7. Repeat steps 3 to 6 for specifed number of epochs
```

```python
def perceptron(X,y,lr, epochs):
  ones_to_merge = np.ones((X.shape[0],1))
  X = np.hstack([X,ones_to_merge])
  w = np.ones(X.shape[1])
  for j in range(epochs):
    for i in range(X.shape[0]):
      y_pred = np.dot(X[i],w)
      y_pred_sign = -1 if y_pred<0 else 1
      if y_pred_sign!=y[i]:
        w = w + X[i]*y[i]*lr


  return w
```

```python
X = np.array([[2, 1],
              [1, -1],
              [-2, -1],
              [-1, 1]])
y = np.array([1, 1, -1, -1])
lr = 0.1
epochs = 10
perceptron(X,y,lr, epochs)
```

```python
ones_to_X = np.ones((X.shape[0],1))
w = perceptron(X,y,lr, epochs)
X = np.hstack([X, ones_to_X])
np.dot(X,w)
```
