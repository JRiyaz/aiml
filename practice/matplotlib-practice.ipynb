```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
from pathlib import Path

path = Path().joinpath("..", "data", "customer_orders.csv")
df = pd.read_csv(path)
df.head()
```

# Histogram

```python
plt.hist(data=df, x="discounted_price", bins=10)
plt.title("Distribution of price")
plt.xlabel("Price")
plt.ylabel("Frequency of count")
plt.show()
```

```python
# Using Seaborn
sns.histplot(data=df, x="discounted_price", bins=100, kde=True)
plt.title("Distribution of price")
plt.xlabel("Price")
plt.ylabel("Frequency of count")
plt.axvline(df["discounted_price"].mean(), label="mean price", color="r")
plt.legend()
```

```python
sns.countplot(data=df, x="category")
plt.title("Count of categories")
plt.xlabel("Category")
plt.ylabel("Frequency of categories")
plt.xticks(rotation=45)
```

```python
category_count = df["category"].value_counts()
plt.bar(category_count.index, category_count.values)
plt.title("Count of categories")
plt.xlabel("Category")
plt.ylabel("Frequency of categories")
plt.xticks(rotation=45)
```

```python
# Using Pandas
category_count.plot(kind="bar")
```

```python
sns.lineplot(data=df, x="category", y="count")
```

# Pairplot

```python
sns.pairplot(df[["discounted_price", "rating", "actual_price", "rating_count"]])
```

# Heatmap

```python
corr_cols = ["discounted_price", "rating", "actual_price", "rating_count"]
corr_matrix = df[corr_cols].corr()
plt.figure(figsize=(5, 5))
sns.heatmap(corr_matrix)
```

1. Categorical variable -> Bar plot
2. Numeric variables    -> Scatter plot
3. Categorical with numerical -> Box plot
4. What parameter is used in Bar plot / Distribution for different categories is `hue`
5. Which plot is used to combine frequency distribution and Box plot? Violin plot
6. Multimodel distribution for different categories? Violin plot
7. Compare correlations HeatMap is used
8. Compare multiple numerical values we used Pairplot
