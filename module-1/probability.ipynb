# Mutually Exclusive Events

Mutually Exclusive Events are the events that can never happen together
e.x
- Scahin scoring a duck (0) and century (100) in same ODI match (The probability of this event is 0 - This will never happen)
- Getting Heads and Tails in same toss (Never happens)

```python
# Find out weather Saching scoring 100 and match ending in tie is mutually exclusive events or not

# df.head()
event = (df['runs'] == 100) & (df['match_result'] == 'tied')
df[event] # As there are no results with the event, It is Mutually Exclusive Event
```

# Exhaustive Events
Exhaustive Events are events (let there are 3 events A, B and C) such that A and B and C make up the entire Sample Space.
e.x
- A. Sachin scoring 100 or more
- B. Sachin scoring less than 100
- C. Sachin did not bat
Note: Here A, B and C are all the possibilties other than these there will no other option.

Rolling a die e.x
- A. All Even numbers
- B. All Odd numbers
NOTE: Both A and B covers all the possible senarios.


**Mutually Exclusive** - Events not possible (No possible outcomes)

**Mutually Exhausting** - The number of events covers all sample space (Common possible but must covers all outcomes)


# Intersection vs Union
PS is pronounced as (P intersection W)
PUS is pronounced as (P union W)

```python
# Get matches in which Sachin scored >= 100 and India won (Intersection)

event = (df["runs"] >= 100) & (df["match_result"] == "won")
all_event = len(df[event])

all_event / len(df)
```

```python
# Get matches in which Sachin scored >= 100 or India won (Union)

event = (df["runs"] >= 100) | (df["match_result"] == "won")
all_event = len(df[event])

all_event / len(df)
```

# Independent Events

Two events are independent when one event has no impact or effect on the other event

e.x,
- Exam results in two different subjects
- Toss a coin and roll a die (Tossing a coin has no effect on rolling a die)

When A and B are independent of each other P(AnB) = P(A) * P(B)

```python
from fractions import Fraction
Fraction(20 / 40)

res = 21 / 420
Fraction(res).limit_denominator()
```

```python
# Check whether it being Monday and Sachin Scoring in single digits are independent events

# [df.groupby("day_of_week")["runs"] < 10].value_counts()
prob_AnB = len(df[(df["day_of_week"] == "Monday") & (df["runs"] < 10)]) / len(df)

prob_A = len(df[df["day_of_week"] == "Monday"]) / len(df) # Probability of Monday

prob_B = len(df[df["runs"] < 10]) / len(df) # Probability of scoring single digit

prod_A_and_prob_B = prob_A * prob_B

print("P(AnB):", prob_AnB)
print("P(A) * P(B):", prod_A_and_prob_B)
print("Does P(AnB) = P(A) * P(B)?", prob_AnB == prod_A_and_prob_B)

print("\nBut as per Butterfly effect no two events are Truly independent")
print("Check this: A flap of butterfly wing in Delhi can cause a tornado in USA")
```

# Conditional Probability
In rolling a die have {1, 2, 3, 4, 5, 6} possible outcomes. The probability of getting 2 is 1/6.

I know beforehand that the outcome will be even number {2, 4, 6}, then the probablility of getting 2 will be 1/3 is called Conditional probability.

Given some condition, what is the probability of some event. In the above example the condition is we know the outcome will be even number.

```python
# Let's say I know Sachin will score 50 or more, what is the probability of India winning.

len(df[(df["runs"] >= 50) & (df["match_result"] == "won")]) / len(df[df["runs"] >= 50])
```
